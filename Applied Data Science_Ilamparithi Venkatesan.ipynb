```python
import numpy as np
import pandas as pd
from plotnine import ggplot, aes, geom_line, geom_bar, geom_point, geom_boxplot
```

**File Reading**

```python
data1=pd.read_csv(r"D:\\Ilamparithi Venkatesan\\Eastbourne Weather.csv", low_memory=False)
```

```python
#Barplot to visualise the minimum temperature of Eastbourne
ggplot(data1) + aes(x="TEMP_MIN_C") + geom_bar()
```

```python
from plotnine import geom_point
```

```python
ggplot(data1) + aes(x="MM", y="RAINFALL_MM") + geom_boxplot()
```

```python
from plotnine import labs
```

```python
#Coloured scatter plot to understand which month has the highest temperature during the timeframe
(
    ggplot(data1)
    + aes(x="Y", y="TEMP_MAX_C", color="MM")
    + labs(
        x="YR",
        y="Highest Temperature in Eastbourne",
        color="Month",
        title="Highest Temperature month wise",
    )
    + geom_point()
)
```
